```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
pca_50 = PCA(n_components=50)
pca_train_50 = pca_50.fit_transform(X_train)
pca_test_50 = pca_50.fit_transform(X_test)
```

```python
k=KNeighborsClassifier(n_neighbors=7)
k.fit(pca_train_50,y_train)
predictions=k.predict(pca_test_50)
predictions
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(y_test, predictions))
```

```python
from sklearn import metrics
print(metrics.accuracy_score(y_test,predictions))
```

```python
rfc=RandomForestClassifier(n_estimators=200)
rfc.fit(pca_train_50,y_train)
pred_rfc=rfc.predict(pca_test_50)
pred_rfc
```

```python
print(metrics.accuracy_score(y_test,pred_rfc))
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import cross_val_score
ada_rf_clf = AdaBoostClassifier(rfc, n_estimators=200, learning_rate=0.5)
accuracies=cross_val_score(estimator=ada_rf_clf,X=pca_train_50,y=y_train,cv=5)
print(accuracies)
```

```python
ada_rf_clf.fit(pca_train_50,y_train)
pred_ada_rf_clf=ada_rf_clf.predict(pca_test_50)
print(metrics.accuracy_score(y_test,pred_ada_rf_clf))
```
