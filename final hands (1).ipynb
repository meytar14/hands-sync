```python
def convert_spontan_df(df):
    l_df=df[df[' Hand Type']=='left']
    r_df=df[df[' Hand Type']=='right']
    new_df=pd.merge(l_df,r_df,on=' Frame ID')
    new_df['diff_posX']=abs(new_df[' Position X_x']-new_df[' Position X_y'])
    new_df['diff_posY']=abs(new_df[' Position Y_x']-new_df[' Position Y_y'])
    new_df['diff_posZ']=abs(new_df[' Position Z_x']-new_df[' Position Z_y'])
    new_df['diff_el_posX']=abs(new_df[' Elbow pos X_x']-new_df[' Elbow pos X_y'])
    new_df['diff_el_posY']=abs(new_df[' Elbow Pos Y_x']-new_df[' Elbow Pos Y_y'])
    new_df['diff_el_posZ']=abs(new_df[' Elbow Pos Z_x']-new_df[' Elbow Pos Z_y'])
    new_df['diff_wr_posX']=abs(new_df[' Wrist Pos X_x']-new_df[' Wrist Pos X_y'])
    new_df['diff_wr_posY']=abs(new_df[' Wrist Pos Y_x']-new_df[' Wrist Pos Y_y'])
    new_df['diff_wr_posZ']=abs(new_df[' Wrist Pos Z_x']-new_df[' Wrist Pos Z_y'])
    
    n_df=pd.DataFrame({'diff_posX':[ new_df['diff_posX'].mean()],
                       'diff_posY':[ new_df['diff_posY'].mean()],
                       'diff_posZ':[ new_df['diff_posZ'].mean()],
                       'diff_el_posX':[ new_df['diff_el_posX'].mean()],
                       'diff_el_posY':[ new_df['diff_el_posY'].mean()],
                       'diff_el_posZ':[ new_df['diff_el_posZ'].mean()],
                       'diff_wr_posX':[ new_df['diff_wr_posX'].mean()],
                       'diff_wr_posY':[ new_df['diff_wr_posY'].mean()],
                       'diff_wr_posZ':[ new_df['diff_wr_posZ'].mean()],
                       'label':[1]
                      })
    return n_df
```

```python
def convert_sync_df(df):
    l_df=df[df[' Hand Type']=='left']
    r_df=df[df[' Hand Type']=='right']
    new_df=pd.merge(l_df,r_df,on=' Frame ID')
    new_df['diff_posX']=abs(new_df[' Position X_x']-new_df[' Position X_y'])
    new_df['diff_posY']=abs(new_df[' Position Y_x']-new_df[' Position Y_y'])
    new_df['diff_posZ']=abs(new_df[' Position Z_x']-new_df[' Position Z_y'])
    new_df['diff_el_posX']=abs(new_df[' Elbow pos X_x']-new_df[' Elbow pos X_y'])
    new_df['diff_el_posY']=abs(new_df[' Elbow Pos Y_x']-new_df[' Elbow Pos Y_y'])
    new_df['diff_el_posZ']=abs(new_df[' Elbow Pos Z_x']-new_df[' Elbow Pos Z_y'])
    new_df['diff_wr_posX']=abs(new_df[' Wrist Pos X_x']-new_df[' Wrist Pos X_y'])
    new_df['diff_wr_posY']=abs(new_df[' Wrist Pos Y_x']-new_df[' Wrist Pos Y_y'])
    new_df['diff_wr_posZ']=abs(new_df[' Wrist Pos Z_x']-new_df[' Wrist Pos Z_y'])
    
    n_df=pd.DataFrame({'diff_posX':[ new_df['diff_posX'].mean()],
                       'diff_posY':[ new_df['diff_posY'].mean()],
                       'diff_posZ':[ new_df['diff_posZ'].mean()],
                       'diff_el_posX':[ new_df['diff_el_posX'].mean()],
                       'diff_el_posY':[ new_df['diff_el_posY'].mean()],
                       'diff_el_posZ':[ new_df['diff_el_posZ'].mean()],
                       'diff_wr_posX':[ new_df['diff_wr_posX'].mean()],
                       'diff_wr_posY':[ new_df['diff_wr_posY'].mean()],
                       'diff_wr_posZ':[ new_df['diff_wr_posZ'].mean()],
                       'label':[2]
                      })
    return n_df
```

```python
def merge_spontan(list_df):
    new_df=pd.DataFrame({'diff_posX':[],'diff_posY':[],'diff_posZ':[],'diff_el_posX':[],'diff_el_posY':[],'diff_el_posZ':[],'diff_wr_posX':[],'diff_wr_posY':[],'diff_wr_posZ':[],'label':[]})
    for i in list_df:
        temp=convert_spontan_df(i)
        new_df=pd.concat([new_df,temp],ignore_index=True)
        
    return new_df
```

```python
def merge_sync(list_df):
    new_df=pd.DataFrame({'diff_posX':[],'diff_posY':[],'diff_posZ':[],'diff_el_posX':[],'diff_el_posY':[],'diff_el_posZ':[],'diff_wr_posX':[],'diff_wr_posY':[],'diff_wr_posZ':[],'label':[]})
    for i in list_df:
        temp=convert_sync_df(i)
        new_df=pd.concat([new_df,temp],ignore_index=True)
        
    return new_df
```

```python
def merge_alone(list_df):
    new_df=pd.DataFrame({'diff_posX':[],'diff_posY':[],'diff_posZ':[],'diff_el_posX':[],'diff_el_posY':[],'diff_el_posZ':[],'diff_wr_posX':[],'diff_wr_posY':[],'diff_wr_posZ':[],'label':[]})
    for i in list_df:
        temp=convert_alone_df(i)
        new_df=pd.concat([new_df,temp],ignore_index=True)
        
    return new_df
```

```python
sp=merge_spontan(spontan_datasets)
al=merge_alone(alone_datasets)
sy=merge_sync(sync_datasets)
train_df=pd.concat([sp,al,sy],ignore_index=True)
train_df
```

```python
spt=merge_spontan(spontan_test_datasets)
alt=merge_alone(alone_test_datasets)
syt=merge_sync(sync_test_datasets)
test_df=pd.concat([spt,alt,syt],ignore_index=True)
test_df
```

```python
X_train = train_df.drop(['label'], axis=1)
y_train = train_df['label']
X_test = test_df.drop(['label'], axis=1)
y_test = test_df['label']
```

# models classification

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report,confusion_matrix
from sklearn import metrics


k=KNeighborsClassifier(n_neighbors=7)
k.fit(X_train,y_train)
predictions=k.predict(X_test)
print(metrics.accuracy_score(y_test,predictions))
```

```python
cm=confusion_matrix(y_test, predictions)
print(cm)
```

```python
axes=sns.heatmap(cm,square=True,annot=True,fmt='d',cbar=True,cmap=plt.cm.GnBu)
axes.set_xlabel('predicted')
axes.set_ylim(0,3.5)
axes.set_ylabel('actual')
```

```python
rfc=RandomForestClassifier(n_estimators=200)
rfc.fit(X_train,y_train)
pred_rfc=rfc.predict(X_test)
print(metrics.accuracy_score(y_test,pred_rfc))
```

```python
cm2=confusion_matrix(y_test, pred_rfc)
axes=sns.heatmap(cm2,square=True,annot=True,fmt='d',cbar=True,cmap=plt.cm.GnBu)
axes.set_xlabel('predicted')
axes.set_ylim(0,3.5)
axes.set_ylabel('actual')
```
